```python
import pandas as pd
import matplotlib.pyplot as plt

# Loading data from a CSV file
df = pd.read_csv('data.csv')

# Display the first few rows of the dataframe to check the data
print(df.head())
fig, ax = plt.subplots(1,2)
boxplot = df.boxplot(column='age', by='sex', ax=ax[0])
# set x label and y label
boxplot.set_xlabel('Giới')
boxplot.set_ylabel('Tuổi')
# delete top title of boxplot
boxplot.set_title('Trung bình tuổi theo giới')

df['age'].plot(kind='box', ax = ax[1])
grouped_stats = df.groupby('sex')['age'].agg(['mean', 'var'])  # 'var' for variance

for i, (sex, sub_df) in enumerate(df.groupby('sex')):
    # Display mean and variance
    y = max(sub_df['age'])  # Position the text above the highest point in the boxplot
    ax[0].text(i+1, y, f'Mean: {grouped_stats.at[sex, "mean"]:.2f}\nStd: {grouped_stats.at[sex, "var"]**(1/2):.2f}',
            horizontalalignment='center', verticalalignment='top')
mean_age = df['age'].mean()
variance_age = df['age'].var()

# Adding annotations for mean and variance on the plot
y = df['age'].max()  # Position the text above the highest point in the boxplot
ax[1].text(1, y, f'Mean: {mean_age:.2f}\nStd: {variance_age**(1/2):.2f}',
        horizontalalignment='center', verticalalignment='top')
plt.title('Trung bình tuổi')
plt.suptitle('')  # Removes the default title for better clarity
ax[1].set_ylabel('')
ax[1].set_xticks([])
ax[1].grid()
plt.savefig('age_distribution.png')
plt.show()
```

```python
# Calculate the mean and variance age group by sex
mean_age = df.groupby('sex')['age'].mean()
var_age = df.groupby('sex')['age'].var()

print("Mean age by sex:")
print(mean_age)

print("\nVariance of age by sex:")
print(var_age)

#
```
